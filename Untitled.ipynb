```python
import pandas as pd
import matplotlib as plt

file_path = "Resources/cities.csv"
df = pd.read_csv(file_path)

df.head()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
df.plot.scatter('Lat','Max Temp')
plt.xlabel('Latitude')
plt.ylabel('Max Temparature (F)')
plt.xlim(-80,100)
plt.ylim(-100,150)
plt.title('City Latitude vs. Max Temperature (04/09/2018)')
plt.grid
plt.savefig('Max Temp.png')
plt.show()
```
